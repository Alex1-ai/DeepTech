#  Importing Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading the dataset

df = pd.read_csv("Business Funding Data.csv", encoding="latin1")
df.head(30)
```

```python
df.info()
```

# Missing Values

```python
# Get total number of entries
len(df)
```

```python
# Determine the percentage of missing values in each column
missing_percentage = df.isnull().sum() / len(df) * 100
missing_percentage
```

```python
# Find unique items in the Effective date column
df["Effective date"].unique()
```

```python
# Find the most occuring date
df["Effective date"].mode()[0]
```

```python
# Fill the missing date with the mode value
df["Effective date"].fillna(df["Effective date"].mode()[0], inplace=True)
df
```

```python
df.info()
```

```python
# Fill the missing values with mean value
df["Investors Count"].fillna(df["Investors Count"].mean(), inplace=True)
df
```

```python
df.info()
```

```python
df["Investors"].unique()
```

```python
# Fill the Investors value with nan with mode value
df["Investors"].fillna(df["Investors"].mode()[0], inplace=True)
df
```

```python
df.info()
```

```python
df["Financing Type"].mode()
```

```python
df["Financing Type"].fillna(df["Financing Type"].mode()[0], inplace=True)
df
```

```python
df.info()
```

```python
df["Financing Type Normalized"].fillna(df["Financing Type Normalized"].mode()[0], inplace=True)
df
```

```python
df.info()
```

# Convert string data to labels

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
```

```python
# List of categorical columns to encode
categorical_cols = ["Investors", "Source Urls", "Categories", "Financing Type Normalized", 'Financing Type', "Website Domain"]

# Applly Label Encoding to each categorical columns
for col in categorical_cols:
    df[col] = label_encoder.fit_transform(df[col])
```
